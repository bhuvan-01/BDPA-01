```python
#Data Cleaning
import pandas as pd
import numpy as np
data = pd.read_csv("diabetic_data.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.drop('encounter_id', axis=1, inplace=True)
```

```python
data.head(10)
```

```python
print("Summary of Missing Value before operations")
print(data.isnull().sum())
```

```python
data.replace('?', np.nan, inplace=True)
```

```python
print("\nSummary of missing values after operation:")
print(data.isnull().sum())
```

```python
data['readmitted'] = data['readmitted'].replace({'<30': 1, '>30': 0, 'NO': 0})
data['readmitted'] = data['readmitted'].astype('bool')
data.head(20)
```

```python
data.dtypes
```

```python
# Calculate percentage of missing values for each column
missing_percentages = (data.isnull().sum() / len(data)) * 100

# Drop columns with more than 90% missing values
columns_to_drop = missing_percentages[missing_percentages > 90].index
data.drop(columns=columns_to_drop, inplace=True)

# Display the DataFrame after dropping columns
print("DataFrame after dropping columns with more than 90% missing values:")
data.head(10)
```

```python
data.shape
```

```python
columns_to_deleted = ['repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 
                     'acetohexamide', 'tolbutamide', 'acarbose', 'miglitol', 
                     'troglitazone', 'tolazamide', 'examide', 'citoglipton', 
                     'glyburide-metformin', 'glipizide-metformin', 
                     'glimepiride-pioglitazone', 'metformin-rosiglitazone', 
                     'metformin-pioglitazone']

# Drop columns
data.drop(columns=columns_to_deleted, inplace=True)

# Display the DataSet after dropping columns
print("DataSet after dropping near zero-variance columns:")
data.head(10)
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.head(10)
```

```python
numerical_cols = ['time_in_hospital',
'num_lab_procedures',
'num_procedures',
'num_medications',
'number_outpatient',
'number_emergency',
'number_inpatient',
'number_diagnoses']
 
Q1 = data[numerical_cols].quantile(0.25)
Q3 = data[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
 
 
filtered_df = data[~((data[numerical_cols] < (Q1 - 1.5 * IQR)) | (data[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
filtered_df.shape
```

```python
output_csv_path = 'cleaned_data.csv'  # You can specify the desired path here

# Write the cleaned DataFrame to a new CSV file
filtered_df.to_csv(output_csv_path, index=False)

print("The cleaned DataFrame has been written to:", output_csv_path)
```
